```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Data
x1 = ['CatBoost', 'BERT-2']
y1 = [.7536, .9807]
x2 = ['SVC', 'BERT-2']
y2 = [.8963, .9807]
# x3 = ['BERT-1', 'BERT-2']
# y3 = [.9701, .9807]

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(5, 3))

# Plot 1
colors = ['orange', 'green']
axs[0].bar(x1, y1, color=colors)
axs[0].set_title('TF-IDF VS BERT', fontsize = 10)
axs[0].axhline(y=y1[0], color='orange', linestyle='--')
axs[0].axhline(y=y1[1], color='green', linestyle='--')
axs[0].text(0, y1[0] + 0.1, f'{(y1[1]-y1[0])* 100:.2f}%', ha='center', color='black')

# Plot 2
axs[1].bar(x2, y2, color=colors)
axs[1].set_title('GLOVE VS BERT' ,fontsize=10)
axs[1].axhline(y=y2[0], color='orange', linestyle='--')
axs[1].axhline(y=y2[1], color='green', linestyle='--')
axs[1].text(0, y2[0] + 0.02, f'{(y2[1]-y2[0])* 100:.2f}%', ha='center', color='black')

# # Plot 3
# axs[2].bar(x3, y3, color=colors)
# axs[2].set_title('Mixed')
# axs[2].axhline(y=y3[0], color='orange', linestyle='--')
# axs[2].axhline(y=y3[1], color='green', linestyle='--')
# axs[2].text(0, y3[0] + 0.02, f'{y3[1]-y3[0]:.4f}', ha='center', color='black')

# Create custom legend handles
custom_handles = [
    Line2D([], [], color='orange', linestyle='--', marker='s', markersize=8),
    Line2D([], [], color='green', linestyle='--', marker='s', markersize=8)
]

# Add legend at the top
legend_labels = ['Traditional', 'BERT']
fig.legend(custom_handles, legend_labels, loc='upper center', 
           bbox_to_anchor=(0.5, 1.10), ncol=2, fancybox=True, shadow=True)

# Add y-axis label
fig.text(0, 0.5, r'$\bf{Accuracy}$', va='center', rotation='vertical', fontsize=10)

# Add x-axis label
fig.text(0.4, 0, r'$\bf{Classifier}$', va='center', rotation='horizontal')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Accuracy values for the models
accuracy_values = [.7471, .7509, .7536, .8963, .8506, .8612, .9701, .9807]

# Model names
model_names = ['SVC', 'XGBoost', 'CatBoost', 'SVC ', 'CatBoost ', 'LogReg', 'BERT-1', 'BERT-2']

# Colors 
colors = ['orange', 'orange', 'orange', 'green', 'green', 'green', 'blue', 'blue']

# Create a bar plot
plt.bar(model_names, accuracy_values, color=colors, width=.6)

# Set the y-axis label
plt.ylabel('Accuracy')
plt.axhline(y=accuracy_values[7], color='blue', linestyle='--')
plt.xticks(model_names, rotation=45, ha='center')
plt.ylim([0, 1])
# Set the title of the plot
# plt.title('Accuracy Comparison of Five Models')

# Create custom legend handles and labels
custom_handles = [plt.Rectangle((0, 0), 1, 1, color='orange'), 
                  plt.Rectangle((0, 0), 1, 1, color='green'),
                  plt.Rectangle((0, 0), 1, 1, color='blue'),
                 ]
legend_labels = ['TF-IDF', 'GLOVE', 'BERT']

# Create a legend at the top
plt.legend(custom_handles, legend_labels, loc='upper center', bbox_to_anchor=(0.5, 1.1),
           ncol=3)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Accuracy values for the five models
accuracy_valuej = [.7431, .7509, .7536]

accuracy_values = [.8963, .8506, .8612]

accuracy_valuem = [.9701, .9807]


# Model names
model_names1 = ['SVC', 'XGBoost', 'CatBoost' ]
model_names2 = ['SVC ', 'CatBoost ', 'LogReg']
model_names3 = ['BERT-1', 'BERT-2']

# Colors 
colors1 = ['orange', 'orange', 'orange']
colors2 = ['green', 'green', 'green']
colors3 = ['blue', 'blue']


# Create a figure and three subplots
fig, axs = plt.subplots(1, 3, figsize=(8, 5))

# Subplot 1
axs[0].bar(model_names1, accuracy_valuej, color=colors1, width=.6)
axs[0].set_ylabel('Accuracy')
# axs[0].set_title('Joker 2019)')
axs[0].set_ylim([0, 1])
axs[0].axhline(y=accuracy_valuej[2], color='black', linestyle='--')

# Subplot 2
axs[1].bar(model_names2, accuracy_values, color=colors2, width=.6)
axs[1].set_ylabel('Accuracy')
# axs[1].set_title('Star Wars Episode VIII')
axs[1].set_ylim([0, 1])
axs[1].axhline(y=accuracy_values[0], color='black', linestyle='--')

# Subplot 3
axs[2].bar(model_names3, accuracy_valuem, color=colors3, width=.5)
axs[2].set_ylabel('Accuracy')
# axs[2].set_title('Mixed')
axs[2].set_ylim([0, 1])
axs[2].axhline(y=accuracy_valuem[1], color='black', linestyle='--')


# Change the rotation of y-axis labels in Subplot 1 to 45 degrees
axs[0].set_xticklabels(model_names1, rotation=45, ha='center')

# Change the rotation of y-axis labels in Subplot 2 to horizontal
axs[1].set_xticklabels(model_names2, rotation=45, ha='center')

# Change the rotation of y-axis labels in Subplot 3 to 45 degrees
axs[2].set_xticklabels(model_names3, rotation=45, ha='center')

# Create custom legend handles and labels
custom_handles = [plt.Rectangle((0, 0), 1, 1, color='orange'), 
                  plt.Rectangle((0, 0), 1, 1, color='green'),
                  plt.Rectangle((0, 0), 1, 1, color='blue'),
                  ]
legend_labels = ['TF-IDF', 'GLOVE', 'BERT']

# Add the legend to the figure
fig.legend(custom_handles, legend_labels, loc='upper center', bbox_to_anchor=(0.5, .98),
           ncol=3)

# Adjust the spacing between subplots and legend
plt.tight_layout(rect=[0, 0, 1, 0.92])


# Show the plot
plt.show();
```
